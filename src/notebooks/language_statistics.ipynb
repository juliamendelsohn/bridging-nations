```python
import os
import sys
import matplotlib
import ujson
import glob
```

```python
relevant_users = {}

with open('/home/sayghosh/cross-lingual-exchange/data/relevant_users.tsv', 'r') as f:
    for user in f:
        relevant_users[user.split("\t")[0].strip()] = {}
```

```python
os.chdir("/shared/1/projects/cross-lingual-exchange/data/relevant_tweets")

for file in glob.glob("*.json"):
    with open(file, 'r') as f:
        print(file)
        all_tweets = ujson.load(f)
        
        for user, tweets in all_tweets.items():
            for tweet in tweets:
                if tweet['lang'] in relevant_users[user]:
                    relevant_users[user][tweet['lang']] += 1
                else:
                    relevant_users[user][tweet['lang']] = 1
```

```python
have_language_data = 0
for user, languages in relevant_users.items():
    if languages:
        have_language_data += 1

print(have_language_data)
```

```python
import pandas as pd
df = pd.read_csv('/home/sayghosh/cross-lingual-exchange/data/relevant_users.tsv', sep='\t', header=None)
print(df.head())
```

```python
df = df.rename(columns={0: 'user_id', 1: 'country'})
print(df.head())
```

```python
df['bilinguality'] = 0
df['num_tweets'] = 0

df.set_index('user_id')
```

```python
import operator

df['bilinguality'] = -1
df['tweets_in_en'] = 0
df['tweets_in_de'] = 0

num_processed = 0

df.set_index('user_id')

for user, languages in relevant_users.items():
    num_processed += 1
    
    if num_processed % 100000 == 0:
        print("{} users processed".format(num_processed))
        
    if 'en' in languages or 'de' in languages:
        # sorted_language_counts = sorted(languages.items(), key=operator.itemgetter(1))
        num_tweets = sum([v for k, v in languages.items() if k == 'en' or k == 'de'])
        
        if len(languages.items()) > 2:
            if 'en' in languages:
                df.loc[user, 'tweets_in_en'] = languages['en']
            if 'de' in languages: 
                df.loc[user, 'tweets_in_de'] = languages['de']
            
            if 'en' in languages and 'de' in languages:
                df.loc[user, 'bilinguality'] = languages['en'] / languages['de']
            
        else:
            if 'en' in languages:
                df.loc[user, 'tweets_in_en'] = languages['en']
                df.loc[user, 'bilinguality'] = 1
            if 'de' in languages:
                df.loc[user, 'tweets_in_en'] = languages['de']
                df.loc[user, 'bilinguality'] = 0
```

```python
df.to_csv(r'/home/sayghosh/cross-lingual-exchange/data/relevant_users_with_language.tsv', index = None, sep='\t', encoding='utf-8')
```
